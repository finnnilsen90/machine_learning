```python
import numpy as np
import matplotlib.pyplot as plt
import math

# generate XY coordinates for a circle
# plot the circle

x = np.linspace(-np.pi,np.pi,100)
xy = np.transpose(np.matrix([np.cos(x), np.sin(x)]))

plt.plot(xy[:,0],xy[:,1],'r','markerfacecolor','w')
plt.grid()
# plt.axis((-6, 6, -6, 6))
plt.title('Circle')
plt.show()
```

```python
# create a 2x2 matrix. Start with an identity matrix.
# multiply the x y coordinates of the circle by the matrix.
# plot the new coordinates
T = np.matrix([[1,0],[1,2]])
newxy = xy*T
plt.plot(xy[:,0],xy[:,1],'r--','markerfacecolor','w')
plt.plot(newxy[:,0],newxy[:,1],'bo','markerfacecolor','m')
plt.grid()
plt.axis((-1, 1, -1, 1)*np.max(np.abs(np.matrix([newxy[:],xy[:]]))))
plt.axis('square')
plt.title('Circle')
plt.show()

# Try with various matricies such as singular matrix. Columns form a linearly dependent set.
```
