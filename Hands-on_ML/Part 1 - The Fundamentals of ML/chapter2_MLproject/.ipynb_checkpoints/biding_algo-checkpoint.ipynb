```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler

channel_data_num = channel_data.drop('Channel', axis=1)
channel_data_num = channel_data_num.drop('Creative Size', axis=1)

class DataFrameSelector(BaseEstimator, TransformerMixin):
    def __init__(self, attribute_names):
        self.attribute_names = attribute_names
    def fit(self, X, y=None):
        return self
    def transform(self, X):
        return X[self.attribute_names].values


num_attribs = list(channel_data_num)

num_pipeline = Pipeline([
    ('selector', DataFrameSelector(num_attribs)),
    ('std_scaler', StandardScaler())
])

channel_data_num_tr = num_pipeline.fit_transform(channel_data_num)
channel_data_num_tr
```

```python
### Standardize the Data ###
# Establish a pipeline
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.base import BaseEstimator, TransformerMixin

num_attribs = list(channel_data_num)
cat_attribs = ['Channel', 'Creative Size']

num_pipeline = Pipeline([
    ('selector', DataFrameSelector(num_attribs)),
    ('std_scaler', StandardScaler())
])

cat_pipeline = Pipeline([
    ('selector', DataFrameSelector(cat_attribs)),
    ('cat_encoder', OneHotEncoder(sparse=False,categories='auto'))
])
```

```python
from sklearn.pipeline import FeatureUnion

full_pipeline = FeatureUnion(transformer_list=[
    ('num_pipeline', num_pipeline),
    ('cat_pipeline', cat_pipeline)
])

channel_data_prepared = full_pipeline.fit_transform(channel_data)
channel_data_prepared
```
